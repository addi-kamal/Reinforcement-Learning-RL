```python
from facebook_scraper import get_posts
import pandas as pd
from datetime import datetime
```

```python
df = pd.DataFrame(columns=['post_id',
    'post_url',
    'text',
    'post_text',
    'time' ,
    'comments' ,
    'shares' ,
    'username',
    'user_url'])
my_keys = ['post_id',
    'post_url',
    'text',
    'post_text',
    'time' ,
    'comments' ,
    'shares' ,
    'username',
    'user_url']
```

```python
%time
for post in get_posts('Mali.Womenonwaves', pages=200):
    post_content = { x: post[x] for x in my_keys }
    df_post = pd.DataFrame.from_dict(post_content, orient='index').T
    df = df.append(df_post)
```

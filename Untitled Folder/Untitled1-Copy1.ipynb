```python
def SignIn(driver, email, password):
    # sign in
    SignIn = driver.find_element_by_xpath('//*[@id="mobile_login_bar"]/div[2]/a[1]')
    SignIn.click()
    email_ = driver.find_element_by_xpath('//*[@id="m_login_email"]')
    email_.send_keys(email)
    password_ = driver.find_element_by_xpath('//*[@id="password_input_with_placeholder"]/input')
    password_.send_keys(password)
    button = driver.find_element_by_xpath('//*[@id="login_form"]/ul/li[3]/input')
    button.click()
    
def get_comments(driver):

    html_source = driver.page_source

    tree = html.fromstring(html_source)
    user_comment = tree.xpath('//*[@id="1027417580772755"]/div/h3/a/text()')
    comment = tree.xpath('//*[@id="1027417580772755"]/div/div[1]/text()')
    like = tree.xpath('//*[@id="like_755810031266846_1027417580772755"]/span/a[1]/text()')
    print(user_comment, comment, like)
        
def view_more_comments(driver):
    more_comments = driver.find_element_by_xpath('//*[@id="see_next_755810031266846"]/a')
    more_comments.click() 
```

```python
def get_comment(i):
    html_source = driver.page_source

    tree = html.fromstring(html_source)
    user_comment = tree.xpath("/html/body/div/div/div[2]/div/div[1]/div/div/div[3]/div[2]/div/div/div[4]/div[{index}]/div/h3/a/text()".format(index=i))
    comment = tree.xpath("/html/body/div/div/div[2]/div/div[1]/div/div/div[3]/div[2]/div/div/div[4]/div[{index}]/div/div[1]/text()".format(index=i))
    like = tree.xpath("/html/body/div/div/div[2]/div/div[1]/div/div/div[3]/div[2]/div/div/div[4]/div[{index}]/div/div[3]/span[1]/span/a[1]/text()".format(index=i))
    # like path change //*[@id="like_{post_id}_1027417580772755"]/span/a[1]/text()
    #                                 post id _???????? 
    lst = user_comment+comment+like 
    return lst
```

```python
def get_comments(driver):
    for i in range(1,11):
        comment_info = get_comment(i)
        d[i] = comment_info
        return d
        
def get_all_comments(driver):
    T = True
    while(T):
        try:
            d = get_comments(driver)
            d.update(d)
            view_more_comments(driver) 
            driver.implicitly_wait(10)
        except:
            T = False
```
